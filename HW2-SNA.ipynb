  ### 1. 소셜 네트워크 분석이란
  
1. 네트워크 분석 방법론과 점점 커지는 중요성       
네트워크 분석 방법론은 속성이 아닌 관계를 중심으로 분석하는 연구방법론으로서 네트워크 관계를 측정하는 다양한 표준화된 척도를 제공한다. 스마트 디바이스의 급격한 확산으로 sns확산이 가속화 되면서 네트워크 분석의 중요성이 커지고있다.     
2. 네트워크의 구성 개념         
네트워크는 간단히 말하면 각각의 객채들이 상호 연결된 구조이다. 노드와 링크로 이루어져있으며, 네트워크는 복수의 노드들이 연결된 구조이다. 간단한 예로는 세계 항공기 노선도와 수도권 지하철 노선도가 있다.    
3. 네트워크 표현 방법                  
방향성, 관계의 정도에 따라서 그래프를 다르게 표현할 수 있으며 인접행렬, 노드리스트, 엣지리스트 방식으로도 그래프를 표현할 수 있다.    
4. 클러스터            
유사한 속성을 가지고 있는 노드들의 집합이다. 노드간 유사성 분석을 통한 클러스터링은 수학적 알고리즘을 기반으로 계산 및 시각화되며 대부분의 소셜네트워크 알고리즘에 내장되어 있어 손쉽게 결과물을 얻을 수 있다.    
5. 소셜 네트워크 분석 지표           
중심성 분석: 가장 많이 사용하는 분석지표로, 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표이다. 중심성 분석을 통해서 전체 네트워크에서 중요한 역할을 하는 노드가 무엇인지 파악 가능하다. 종류에 따라서 연결정도 중심성, 근접 중심성, 매개 중심성, 아이겐 벡터 중심성이 있다.   
   




### 2. 네트워크 분석 코딩

```r
### 0. 데이터 불러오기
data <- read.csv("Society Data2.csv"); data <- data[,-1] #분석에 사용하지 않을 컬럼 삭제
str(data) #데이터를 확인
summary(data)
# 데이터를 불러오는 과정에서 변형된 변수속성 다시 factor로 변환
for (i in 9:24) {
  data[,i] <- as.factor(data[,i])
} 


### 1. 연관분석 
# 연관분석 할 데이터 정제 
adata <- data; adata <-adata[,-c(1,3:8,12:13)] #분석에 사용하지 않을 컬럼 제외 
# 범주형으로 데이터 정제
adata$일자리만족도 <- factor(adata$일자리만족도, levels=c(1:5),
                       labels=c('매우만족', '약간만족', '보통', '약간불만족', '매우불만족'))

# 그래프 고정
set.seed(1234)

# 사용할 패키지 불러오기 
library(arules)
library(arulesViz)
rule <- apriori(adata, parameter=list(minlen=2)) #처음 코드를 실행할 때 support=0.025, confidence=0.2 로 실행해보았으나 연관규칙이 너무 많이 나와서 디폴트값으로 코드를 실행하였습니다

# 네트워크 시각화 
plot(rule, method = "graph")

### 2. 원래 데이터의 네트워크
# 원래 데이터의 네트워크 
tbl_data <- as_tbl_graph(data)
tbl_data # 14715 노드와 14713 엣지(링크)로 이뤄졌습니다 
# 시각화 
plot(tbl_data)
```
